Here, because there are some missing values in the "peeks_on_chart" and "Peak_position" columns, we create a cleaned version of the "billboardHot100_1999-2019" table by calculating these metrics on our own. This ensures that the original data can be analyzed properly to capture song performance.

```python
%%bigquery
CREATE OR REPLACE TABLE `a02ba775fall24.cleaned.billboardHot100_1999-2019` AS
WITH DataImpute AS (
  SELECT
    Artists,
    Name,
    COUNT(Weekly_rank) AS Weeks_on_chart,
    MIN(Weekly_rank) AS Peak_position
  FROM
    `a02ba775fall24.billboard.billboardHot100_1999-2019`
  GROUP BY
    Artists,
    Name
)
SELECT
  a.Artists,
  a.Name,
  a.Week,
  a.Date,
  a.Weekly_rank,
  b.Weeks_on_chart,
  b.Peak_position,
  a.Genre,
  ARRAY_LENGTH(SPLIT(a.Genre, ',')) AS genre_count,
  a.features,
  ARRAY_LENGTH(SPLIT(a.features, ',')) AS features_count
FROM
  `a02ba775fall24.billboard.billboardHot100_1999-2019` AS a
LEFT JOIN
  DataImpute AS b
ON
  a.Artists = b.Artists AND
  a.Name = b.Name;
```

This query cleans multiple tables by creating distinct versions of the artistDf, riaaSingleCerts_1999-2019, and grammySongs_1999-2019 tables. It standardizes missing data (like Genres and Group_Solo in artistDf, and Artist in grammySongs) while retaining only unique rows to ensure data consistency for further analysis.

```python
%%bigquery

WITH CleanedArtist AS (
  SELECT DISTINCT
    Artist,
    Followers,
    IFNULL(Genres, 'Unknown') AS Genres,
    NumAlbums,
    YearFirstAlbum,
    Gender,
    IFNULL('Group.Solo', 'Unknown') AS Group_Solo
  FROM
    `a02ba775fall24.billboard.artistDf`
  WHERE
    Gender IS NOT NULL
)
SELECT * FROM CleanedArtist;

WITH CleanedRIAASingles AS (
  SELECT DISTINCT
    Name AS Song_Name,
    Artist,
    RiaaStatus,
    Label
  FROM
    `a02ba775fall24.billboard.riaaSingleCerts_1999-2019`
)
SELECT * FROM CleanedRIAASingles;

WITH CleanedGrammySongs AS (
  SELECT DISTINCT
    *,
    NULLIF(IFNULL(Artist, 'Various Artists'), '') AS Cleaned_Artist
  FROM
    `a02ba775fall24.billboard.grammySongs_1999-2019`
)
SELECT *
FROM CleanedGrammySongs
WHERE Cleaned_Artist IS NOT NULL
ORDER BY Cleaned_Artist ASC;
```

We removed duplicate rows from all three tables to ensure unique entries.
1. Missing values in key columns like Genres, Group.Solo, and Artist were replaced with appropriate placeholders ('Unknown' or 'Various Artists').
2. Rows with missing gender information were filtered out to maintain data completeness.

We look for the most-followed artist in each category: Male Solo, Male Group, Female Solo, and Female Group.
By categorizing and filtering based on Gender and Group_Solo status, it identifies the top artist with the highest follower count within each category.

```python
%%bigquery
WITH Male_Solo AS (
    SELECT
        'Male Solo' AS Category,
        Artist,
        Followers
    FROM
        `a02ba775fall24.billboard.artistDf`
    WHERE
        Gender = 'M' AND Group_Solo = 'Solo'
    ORDER BY
        Followers DESC
    LIMIT 1
),
Male_Group AS (
    SELECT
        'Male Group' AS Category,
        Artist,
        Followers
    FROM
        `a02ba775fall24.billboard.artistDf`
    WHERE
        Gender = 'M' AND Group_Solo = 'Group'
    ORDER BY
        Followers DESC
    LIMIT 1
),
Female_Solo AS (
    SELECT
        'Female Solo' AS Category,
        Artist,
        Followers
    FROM
        `a02ba775fall24.billboard.artistDf`
    WHERE
        Gender = 'F' AND Group_Solo = 'Solo'
    ORDER BY
        Followers DESC
    LIMIT 1
),
Female_Group AS (
    SELECT
        'Female Group' AS Category,
        Artist,
        Followers
    FROM
        `a02ba775fall24.billboard.artistDf`
    WHERE
        Gender = 'F' AND Group_Solo = 'Group'
    ORDER BY
        Followers DESC
    LIMIT 1
)

SELECT * FROM Male_Solo
UNION ALL
SELECT * FROM Male_Group
UNION ALL
SELECT * FROM Female_Solo
UNION ALL
SELECT * FROM Female_Group;
```

Having uncovered these insights from the Billboard chart, we now expand our exploration to assess how commercial success aligns with critical acclaim by analyzing Grammy wins, platinum albums, and certified singles. This expanded exploration offers a holistic view of artist impact in the music industry.

```python
%%bigquery
SELECT
    a.Artist,
    COUNT(DISTINCT b.Name) AS BillboardHits,
    COUNT(DISTINCT g.Album) AS GrammyWins,
    COUNT(DISTINCT r.Album) AS PlatinumAlbums,
    COUNT(DISTINCT s.Name) AS CertifiedSingles
FROM `a02ba775fall24.billboard.artistDf` a
LEFT JOIN `a02ba775fall24.billboard.billboardHot100_1999-2019` b ON a.Artist = b.Artists
LEFT JOIN `a02ba775fall24.billboard.grammyAlbums_1999-2019` g ON a.Artist = g.Artist
LEFT JOIN `a02ba775fall24.billboard.riaaAlbumCerts_cleaned` r ON a.Artist = r.Artist
LEFT JOIN `a02ba775fall24.billboard.riaaSingleCerts_1999-2019` s ON a.Artist = s.Artist
GROUP BY a.Artist
ORDER BY BillboardHits DESC, GrammyWins DESC, PlatinumAlbums DESC, CertifiedSingles DESC
LIMIT 10;
```

The analysis reveals that the Glee Cast recorded an impressive 196 Billboard hits but received zero awards, highlighting a lack of recognition despite chart presence. In contrast, Drake and Taylor Swift demonstrate a balance between commercial success and critical recognition, achieving both high Billboard hits and numerous certified singles.

We will now delve into recent Grammy Award winners across major categories. This analysis focuses on artists and tracks that have achieved critical acclaim, providing insights into the intersection of popularity and artistic recognition.

```python
%%bigquery
SELECT GrammyAward, GrammyYear, Name AS Song_Name, Artist
FROM (
    SELECT Award AS GrammyAward, GrammyYear, Genre, Album AS Name, Artist
    FROM `a02ba775fall24.billboard.grammyAlbums_1999-2019`

    UNION ALL

    SELECT GrammyAward, GrammyYear, Genre, Name, Artist
    FROM `a02ba775fall24.billboard.grammySongs_1999-2019`
) AS CombinedData
WHERE GrammyAward IN ('Best R&B Song','Best Rap Song','Best Pop Song','Song of the Year','Record of the Year',
    'Best Pop Album','Best R&B Album','Best Rap Album','Best Rock Album','Album Of The Year','Best New Artist')
    AND
    2019 - GrammyYear < 5
ORDER BY GrammyYear ASC;
```


To further understand the relationship between critical acclaim and artistic achievement, we analyze the artists with the most Grammy wins. This highlights the top Grammy winners, providing a clearer picture of those who have consistently demonstrated excellence and influence in the music industry.




```python
%%bigquery
SELECT Artist, COUNT(*) AS Number_of_Grammy
FROM (
    SELECT Award AS GrammyAward, GrammyYear, Genre, Album AS Name, Artist
    FROM `a02ba775fall24.billboard.grammyAlbums_1999-2019`

    UNION ALL

    SELECT GrammyAward, GrammyYear, Genre, Name, Artist
    FROM `a02ba775fall24.billboard.grammySongs_1999-2019`
) AS CombinedData
WHERE Artist IS NOT NULL AND Artist <> 'Various Artists' AND  TRIM(Artist) <> ''
GROUP BY Artist
ORDER BY Number_of_Grammy DESC
LIMIT 15;
```

Adele leads with 12 Grammy wins, showcasing her unparalleled critical acclaim, followed closely by Eminem with 11 and Alicia Keys with 10. These artists exemplify consistent excellence and influence in shaping the music landscape.

```python
%%bigquery
SELECT
    certification_level,
     CAST(AVG(avg_streams) AS INT64) AS avg_streams_per_certification
FROM (
    SELECT
        s.Name AS song_name,
        s.Artist AS artist_name,
        AVG(s.Streams) AS avg_streams,
        CASE
            WHEN r.RiaaStatus LIKE '%Diamond%' THEN 'Diamond'
            WHEN r.RiaaStatus LIKE '%Multi-Platinum%' THEN 'Multi-Platinum'
            WHEN r.RiaaStatus LIKE '%Platinum%' THEN 'Platinum'
            WHEN r.RiaaStatus LIKE '%Gold%' THEN 'Gold'
            ELSE 'No Certification'
        END AS certification_level
    FROM
        `a02ba775fall24.billboard.spotifyWeeklyTop200Streams` AS s
    LEFT JOIN
        `a02ba775fall24.billboard.riaaSingleCerts_1999-2019` AS r
        ON s.Name = r.Name AND s.Artist = r.Artist
    GROUP BY
        artist_name, song_name, certification_level
)
GROUP BY
    certification_level
ORDER BY
    avg_streams_per_certification DESC;
#move to indepth
```

This analysis examines the evolution of music genre popularity on the Billboard Hot 100 over four distinct time periods: 1999-2004, 2005-2008, 2009-2014, and 2015-2019.

```python
%%bigquery

WITH GenreCounts AS (
    SELECT
        year_group,
        Genre,
        COUNT(*) AS Genre_Count
    FROM (
        SELECT DISTINCT
            LOWER(TRIM(Name)) AS Name,
            LOWER(TRIM(Artists)) AS Artists,
            EXTRACT(YEAR FROM Week) AS Year,
            CASE
                WHEN EXTRACT(YEAR FROM Week) BETWEEN 1999 AND 2004 THEN '1999-2004'
                WHEN EXTRACT(YEAR FROM Week) BETWEEN 2005 AND 2008 THEN '2005-2008'
                WHEN EXTRACT(YEAR FROM Week) BETWEEN 2009 AND 2014 THEN '2009-2014'
                WHEN EXTRACT(YEAR FROM Week) BETWEEN 2015 AND 2019 THEN '2015-2019'
                ELSE 'Other'
            END AS year_group,
            Genre
        FROM `a02ba775fall24.billboard.billboardHot100_1999-2019`,
        UNNEST(SPLIT(Genre, ',')) AS Genre
        GROUP BY Name, Artists, Year, year_group, Genre
    ) AS Subquery
    GROUP BY year_group, Genre
),
RankedGenres AS (
    SELECT
        year_group,
        Genre,
        Genre_Count,
        ROW_NUMBER() OVER (
            PARTITION BY year_group
            ORDER BY Genre_Count DESC
        ) AS rn
    FROM GenreCounts
)
SELECT year_group, Genre, Genre_Count
FROM RankedGenres
WHERE rn <= 5
ORDER BY year_group, Genre_Count DESC;
```

The chart shows that Pop maintained the highest popularity for a significant period but lost its top position to Rap between 2015 and 2019. In contrast, Country and R&B consistently remain in the lower rankings throughout the timeline.

```python
%%bigquery

WITH
OVERALL_TEMP AS (
  SELECT
    ROUND(100 * COUNT(DISTINCT CASE WHEN Explicit = True THEN Name ELSE NULL END)
          / COUNT(DISTINCT Name), 4) AS overall_exp_per
  FROM `a02ba775fall24.billboard.songAttributes_1999-2019`
  WHERE Explicit IS NOT NULL
),
GRAMMY_TEMP AS (
  WITH GRAMMYSONG AS (
    SELECT
        GrammyAward,
        GrammyYear,
        LOWER(Name) AS Name,
        LOWER(Artist) AS Artist
    FROM `a02ba775fall24.billboard.grammySongs_1999-2019`
  ),
  SONGATTRIBUTE AS (
    SELECT
        LOWER(Name) AS Name,
        LOWER(Artist) AS Artist,
        Explicit
    FROM `a02ba775fall24.billboard.songAttributes_1999-2019`
  )
  SELECT
      DISTINCT a.Name AS OriginalName,
      a.Artist AS OriginalArtist,
      a.GrammyAward,
      a.GrammyYear,
      b.Explicit
  FROM GRAMMYSONG AS a
  LEFT JOIN SONGATTRIBUTE AS b
  ON a.Name = b.Name
  AND a.Artist = b.Artist
  WHERE Explicit IS NOT NULL
),
RIAA_TEMP AS (
  WITH RIAASONG AS (
    SELECT
        RiaaStatus,
        LOWER(Name) AS Name,
        LOWER(Artist) AS Artist
    FROM `a02ba775fall24.billboard.riaaSingleCerts_1999-2019`
  ),
  SONGATTRIBUTE AS (
     SELECT
         LOWER(Name) AS Name,
         LOWER(Artist) AS Artist,
         Explicit
     FROM `a02ba775fall24.billboard.songAttributes_1999-2019`
  )
  SELECT
      DISTINCT a.Name AS OriginalName,
      a.Artist AS OriginalArtist,
      a.RiaaStatus,
      b.Explicit
  FROM RIAASONG AS a
  LEFT JOIN SONGATTRIBUTE AS b
  ON a.Name = b.Name
  AND a.Artist = b.Artist
  WHERE b.Explicit IS NOT NULL
)
SELECT
  (SELECT overall_exp_per FROM OVERALL_TEMP) AS overall_exp_per,
  ROUND(100 * COUNT(DISTINCT CASE WHEN gram.Explicit = True THEN gram.OriginalName ELSE NULL END)
        / COUNT(DISTINCT gram.OriginalName), 4) AS gram_exp_perc,
  ROUND(100 * COUNT(DISTINCT CASE WHEN riaa.Explicit = True THEN riaa.OriginalName ELSE NULL END)
        / COUNT(DISTINCT riaa.OriginalName), 4) AS riaa_exp_perc
FROM GRAMMY_TEMP AS gram, RIAA_TEMP AS riaa
WHERE gram.Explicit IS NOT NULL
AND riaa.Explicit IS NOT NULL;
```

Our results indicate that 23.77% of all songs in the dataset are explicit. Among Grammy-winning songs, however, only 13.97% are explicit, compared to 23.01% of RIAA-certified songs, which closely aligns with the overall percentage. While explicit content may have a neutral impact on commercial success, it may be less favorable for critical recognition, as seen in the lower prevalence of explicit songs among Grammy winners.

```python
%%bigquery

CREATE OR REPLACE VIEW `a02ba775fall24.billboard.billboard_explicit_popularity_correlations` AS
WITH BILLBOARDSONG AS (
    SELECT
        Peak_position,
        Weeks_on_chart,
        LOWER(Name) AS Name,
        LOWER(Artists) AS Artist,
        SAFE_CAST(TRIM(Weeks_on_chart) AS INT64) AS WeeksOnChartInt,
        ROW_NUMBER() OVER (
            PARTITION BY LOWER(Name), LOWER(Artists)
            ORDER BY SAFE_CAST(TRIM(Weeks_on_chart) AS INT64) DESC
        ) AS rn_weeks
    FROM `a02ba775fall24.billboard.billboardHot100_1999-2019`
    WHERE SAFE_CAST(TRIM(Weeks_on_chart) AS INT64) IS NOT NULL
),
SONGATTRIBUTE AS (
    SELECT
        LOWER(Name) AS Name,
        LOWER(Artist) AS Artist,
        Explicit,
        Popularity,
        ROW_NUMBER() OVER (
            PARTITION BY LOWER(Name), LOWER(Artist), Explicit
            ORDER BY Popularity DESC
        ) AS rn_popularity
    FROM `a02ba775fall24.billboard.songAttributes_1999-2019`
    WHERE Explicit IS NOT NULL
)
SELECT
    a.Name AS OriginalName,
    a.Artist AS OriginalArtist,
    a.Peak_position,
    a.Weeks_on_chart,
    a.WeeksOnChartInt,
    b.Explicit,
    b.Popularity
FROM BILLBOARDSONG AS a
LEFT JOIN SONGATTRIBUTE AS b
    ON a.Name = b.Name
    AND a.Artist = b.Artist
WHERE a.rn_weeks = 1
  AND b.rn_popularity = 1
ORDER BY OriginalName DESC;
```

```python
%%bigquery
WITH Base AS (
  SELECT
    DATE_TRUNC(Date, YEAR) AS Year,
    genre
  FROM
    `a02ba775fall24.billboard.billboardHot100_1999-2019`,
    UNNEST(SPLIT(genre, ',')) AS genre
  WHERE DATE IS NOT NULL
),
Base_2 AS (
  SELECT
    Year,
    genre,
    COUNT(Year) AS count
  FROM
    Base
  GROUP BY
    Year, genre
),
Base_3 AS (
  SELECT
    Year,
    genre,
    count,
    RANK() OVER (PARTITION BY Year ORDER BY count DESC) AS rank
  FROM
    Base_2
)
SELECT
  Year,
  STRING_AGG(genre, ', ') AS top_genres
FROM
  Base_3
WHERE
  rank in (1,2,3,4,5)
GROUP BY
  Year
ORDER BY
  Year;
```

```python
%%bigquery
WITH Spotify_ArtistDf AS (
  SELECT
    a.Name,
    a.Artist,
    a.Streams,
    DATE_TRUNC(a.Week, MONTH) AS Month,
    genre
  FROM
    `a02ba775fall24.billboard.spotifyWeeklyTop200Streams` a
    LEFT JOIN `a02ba775fall24.billboard.artistDf` b
    ON LOWER(TRIM(a.Artist)) = LOWER(TRIM(b.Artist))
    CROSS JOIN UNNEST(SPLIT(b.Genres, ',')) AS genre
  WHERE
    b.Genres IS NOT NULL
),
Spotify_ArtistDf_2 AS (
  SELECT
    Month,
    genre,
    SUM(Streams) AS SUM_Streams,
    RANK() OVER (PARTITION BY Month ORDER BY SUM(Streams) DESC) AS rank
  FROM
    Spotify_ArtistDf
  GROUP BY
    Month, genre
)
SELECT
  Month,
  genre,
  SUM_Streams,
  rank
FROM
  Spotify_ArtistDf_2
WHERE
  rank IN (1, 2, 3, 4, 5) AND Month BETWEEN "2017-01-01" AND "2019-06-01"
ORDER BY
  Month, rank;
```

Now, we would like to explore the relationship between RIAA certification diversity and streaming success among top artists. By examining each artist's genre, streaming performance, and range of certifications, we identify key players by joining the artistDF and spotifyWeeklyTop200Stream datasets. We then order the results by the count of certifications and average streams.

```python
%%bigquery
SELECT
    a.Artist,
    FORMAT('%.0f', AVG(s.Streams)) AS avg_weekly_streams,
    COUNT(DISTINCT r.RiaaStatus) AS certification_count,
    ARRAY_AGG(DISTINCT r.RiaaStatus IGNORE NULLS) AS certification_types
FROM
    `a02ba775fall24.billboard.artistDf` AS a
JOIN
    `a02ba775fall24.billboard.spotifyWeeklyTop200Streams` AS s ON a.Artist = s.Artist
LEFT JOIN
    `a02ba775fall24.billboard.riaaSingleCerts_1999-2019` AS r ON a.Artist = r.Artist AND s.Name = r.Name
GROUP BY
    a.Artist
ORDER BY
    certification_count DESC, avg_weekly_streams DESC
LIMIT 10;
```

The results highlight the top artists based on their diversity in RIAA certifications and average streaming performance. Below are detailed insights into the top three artists:

Moving on, we aim to investigate the relationship between commercial success, measured by RIAA certifications, and critical recognition, measured by Grammy Awards, for prominent artists. By focusing on artists with multiple RIAA-certified songs (excluding Gold level, to highlight higher certification tiers), we explore patterns in both popularity and acclaim, examining how often high certification levels correspond with Grammy recognition.

We introduce a new column - RIAAtoGrammyRatio

A higher value of this ratio indicates an artist whose songs have been commercially successful but have not received as many Grammy awards.
This suggests that while an artist's music is highly popular and widely sold, it may not always receive critical acclaim.

```python
#Discuss to convert 'RiaaStatuses' to RiaaStatuses (score by weight)
%%bigquery
WITH GrammyWins AS (
  SELECT
    Artist,
    COUNT(*) AS NumGrammyAwards
  FROM
    `a02ba775fall24.billboard.grammySongs_1999-2019`
  WHERE
    Artist IS NOT NULL
  GROUP BY
    Artist
),
RIAAGrammySongs AS (
  SELECT
    r.Name AS SongName,
    r.Artist AS ArtistName,
    r.RiaaStatus,
    g.GrammyAward,
    g.GrammyYear,
    a.NumAlbums,
    IFNULL(w.NumGrammyAwards, 0) AS NumGrammyAwards,
    g.Genre
  FROM
    `a02ba775fall24.billboard.riaaSingleCerts_1999-2019` r
  LEFT JOIN
    `a02ba775fall24.billboard.grammySongs_1999-2019` g
  ON
    r.Name = g.Name AND r.Artist = g.Artist
  JOIN
    `a02ba775fall24.billboard.artistDf` a
  ON
    r.Artist = a.Artist
  LEFT JOIN
    GrammyWins w
  ON
    r.Artist = w.Artist
  WHERE
    r.RiaaStatus != 'Gold' -- Exclude Gold RIAA certifications
),
AggregatedRIAAGrammySongs AS (
  SELECT
    ArtistName,
    STRING_AGG(DISTINCT Genre, ', ') AS Genres,
    STRING_AGG(DISTINCT RiaaStatus, ', ') AS RiaaStatuses,
    COUNT(DISTINCT SongName) AS NumRIAACertifiedSongs,
    MAX(NumGrammyAwards) AS NumGrammyAwards
  FROM
    RIAAGrammySongs
  GROUP BY
    ArtistName
  HAVING
    NumGrammyAwards > 0
),
FinalResult AS (
  SELECT
    ArtistName,
    Genres,
    RiaaStatuses,
    NumRIAACertifiedSongs,
    NumGrammyAwards,
    SAFE_DIVIDE(NumRIAACertifiedSongs, NumGrammyAwards) AS RIAAtoGrammyRatio
  FROM
    AggregatedRIAAGrammySongs
)
SELECT * FROM FinalResult
ORDER BY NumRIAACertifiedSongs DESC
LIMIT 10;
```

The results reveal interesting insights into the relationship between commercial success and Grammy recognition for top artists.
Drake, with a remarkable 69 RIAA-certified songs and 3 Grammy Awards, has a high RIAA-to-Grammy ratio of 23, indicating that his commercial success far outweighs his Grammy accolades. This suggests that while Drake’s music has wide-reaching popularity and sales, it may not have achieved the same level of critical acclaim from the Grammy Awards.

Similarly, Taylor Swift has 41 RIAA-certified songs and 4 Grammy Awards, resulting in a ratio of 10.25. Though she has received critical recognition, the data hints that her commercial success—especially in the country and pop genres—outpaces her Grammy wins.

To move forward we’re now analyzing how Grammy Award recognition has evolved for different music genres across two decades. By comparing the number of awards won in the 2000s and 2010s, we identify which genres have gained or lost prominence and classify these shifts as increases, decreases, or no change. This insight highlights trends in musical preferences and industry priorities, providing a strategic perspective on evolving audience and critical acclaim dynamics.

```python
%%bigquery
WITH Decade2000s AS (
    SELECT
        Genre,
        COUNT(DISTINCT Award) AS Awards2000s
    FROM
        `a02ba775fall24.billboard.grammyAlbums_1999-2019`
    WHERE
        GrammyYear BETWEEN 2000 AND 2009
    GROUP BY
        Genre
),
Decade2010s AS (
    SELECT
        Genre,
        COUNT(DISTINCT Award) AS Awards2010s
    FROM
        `a02ba775fall24.billboard.grammyAlbums_1999-2019`
    WHERE
        GrammyYear BETWEEN 2010 AND 2019
    GROUP BY
        Genre
)
SELECT
    COALESCE(d2k.Genre, d2t.Genre) AS Genre,
    IFNULL(d2k.Awards2000s, 0) AS Awards2000s,
    IFNULL(d2t.Awards2010s, 0) AS Awards2010s,
    IFNULL(d2t.Awards2010s, 0) - IFNULL(d2k.Awards2000s, 0) AS AwardChange,
    CASE
        WHEN IFNULL(d2t.Awards2010s, 0) - IFNULL(d2k.Awards2000s, 0) > 0 THEN 'Increase'
        WHEN IFNULL(d2t.Awards2010s, 0) - IFNULL(d2k.Awards2000s, 0) < 0 THEN 'Decrease'
        ELSE 'No Change'
    END AS ChangeType
FROM
    Decade2000s d2k
FULL OUTER JOIN
    Decade2010s d2t
ON
    d2k.Genre = d2t.Genre
ORDER BY
    ABS(AwardChange) DESC,
    AwardChange DESC,
    Awards2000s DESC,
    Awards2010s DESC

LIMIT 10;
```

The analysis reveals key shifts in Grammy recognition across genres, offering insights into changing industry and audience priorities:



**Genres in Decline:** Traditional categories like Latin (-4) and Gospel/Contemporary Christian Music (-2) have seen reduced recognition, potentially reflecting a shift in focus away from these genres in award ceremonies or broader market trends.

**Emerging Genres:** Dance/Electronic Music (+2) and Contemporary Instrumental Music (+1) show increasing recognition, indicating growing interest and influence in these modern, evolving genres.

**Consistent Growth:** American Roots Music (+2), Jazz (+1), and Classical (+1) continue to garner steady acclaim, suggesting their enduring relevance and a dedicated audience base.

**Opportunities in Niche Areas:** Emerging categories like Musical Theater (+1) and Production, Immersive Audio (+1) demonstrate potential for expansion as they gain recognition in smaller increments.

This data can guide investment in rising genres and indicate areas where declining categories may need strategic revitalization to maintain relevance.

This analysis evaluates artist success across Billboard hits, Grammy wins, and RIAA certifications by introducing the "Total RIAA Certification Weight." This weight assigns proportional values to certifications, with higher significance given to Diamond and multi-Platinum achievements, offering a nuanced measure of commercial impact.

The results help highlight potential anomalies, such as artists with exceptional RIAA certification weights but limited Grammy wins or chart success, revealing divergences between commercial popularity and critical acclaim. These insights allow record labels and music producers to better understand the dynamics of audience preferences, critical recognition, and revenue potential, shaping more effective strategies in artist development and music production.

```python
%%bigquery

WITH RIAACertifications AS (
    SELECT
        Artist,
        Album,
        SUM(
            CASE
                WHEN Status LIKE '%Gold%' THEN 0.5
                WHEN Status LIKE '%Platinum%' THEN COALESCE(CAST(REGEXP_EXTRACT(Status, r'(\\d+)x') AS FLOAT64), 1)
                WHEN Status LIKE '%Diamond%' THEN COALESCE(CAST(REGEXP_EXTRACT(Status, r'(\\d+)x') AS FLOAT64), 1) * 10
                ELSE 0
            END
        ) AS CertificationWeight
    FROM `a02ba775fall24.billboard.riaaAlbumCerts_cleaned`
    GROUP BY Artist, Album
)
SELECT
    a.Artist,
    COUNT(DISTINCT b.Name) AS BillboardHits,
    COUNT(DISTINCT g.Album) AS GrammyWins,
    SUM(rc.CertificationWeight) AS TotalRIAACertificationWeight
FROM `a02ba775fall24.billboard.artistDf` a
LEFT JOIN `a02ba775fall24.billboard.billboardHot100_1999-2019` b ON LOWER(a.Artist) = LOWER(b.Artists)
LEFT JOIN `a02ba775fall24.billboard.grammyAlbums_1999-2019` g ON LOWER(a.Artist) = LOWER(g.Artist)
LEFT JOIN RIAACertifications rc ON LOWER(a.Artist) = LOWER(rc.Artist)
GROUP BY a.Artist
ORDER BY GrammyWins DESC, BillboardHits DESC, TotalRIAACertificationWeight DESC
LIMIT 10;
```

This chart showcases the diverse factors contributing to artist success and highlights key strategic opportunities:

**Cross-Segment Dominance**: Eminem stands out with a rare combination of massive commercial success (51 Billboard hits) and high RIAA certification weight (79,542.0), indicating unparalleled cultural and commercial impact.

**Critical Acclaim Specialists**: Artists like Tony Bennett and Kirk Franklin demonstrate significant Grammy recognition (6 and 4 wins, respectively) despite minimal Billboard presence, suggesting strong appeal in niche markets and dedicated fan bases.

**Balanced Success**: Kanye West, Coldplay, and Shakira illustrate balanced achievements across Billboard hits, Grammy wins, and RIAA certifications, making them well-rounded investments with both critical and commercial appeal.

**Genre-Specific Impact**: The Dixie Chicks and Randy Travis highlight country music's unique audience dynamics, with certifications and Grammys reflecting genre loyalty despite relatively fewer Billboard hits.

This data can guide targeted investments, marketing strategies, and catalog curation by aligning with the strengths of top artists—whether it's leveraging Eminem's global appeal or nurturing niche audiences for artists like Tony Bennett.
